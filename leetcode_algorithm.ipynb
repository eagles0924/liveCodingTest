```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        i = 0; k = len(nums) - 1; j = i + 1
        ans = []
        while i < j < k:
            total = nums[i] + nums[j] + nums[j]
            if total == 0:
                ans.append([nums[i], nums[j], nums[k]])
            else:
                if nums[i] > nums[k]:
                    i += 1
                else:
                    k -= 1
```

## 168. Two Sum II

```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        left = 0; right = len(numbers) - 1
        while left < right:
            current_sum = numbers[left] + numbers[right]
            if current_sum == target:
                return [left + 1, right + 1]
            elif current_sum < target:
                left += 1
            else:
                right -= 1
        return []
```

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = s.lower()
        test = []
        for txt in s:
            if txt.isalnum() == True:
                test.append(txt)
        if not test:
            return True
        test = ''.join(test)
        left = 0; right = len(test) - 1
        while left < right:
            if test[left] == test[right]:
                left += 1; right -= 1
            else:
                return False
        return True
s = "A man, a plan, a canal: Panama"
sol = Solution()
sol.isPalindrome(s)
```
