## 3Sum

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        i = 0; k = len(nums) - 1; j = i + 1
        ans = []
        while i < j < k:
            total = nums[i] + nums[j] + nums[j]
            if total == 0:
                ans.append([nums[i], nums[j], nums[k]])
            else:
                if nums[i] > nums[k]:
                    i += 1
                else:
                    k -= 1
```

```python
# nums[i] == nums[i-1] vs nums[i] == nums[i+1] 차이점 비교

def threeSum_skip_previous(nums):
    """현재 방식: nums[i] == nums[i-1]인 경우 skip"""
    nums.sort()
    result = []
    n = len(nums)
    
    print("=== 방식 1: nums[i] == nums[i-1] ===")
    print(f"배열: {nums}")
    print("처리 순서:")
    
    for i in range(n - 2):
        if i > 0 and nums[i] == nums[i - 1]:
            print(f"  i={i}: nums[{i}]={nums[i]} == nums[{i-1}]={nums[i-1]} → SKIP")
            continue
            
        print(f"  i={i}: nums[{i}]={nums[i]} → PROCESS")
        
        # 실제 triplet 찾기 (간단화)
        left, right = i + 1, n - 1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total == 0:
                result.append([nums[i], nums[left], nums[right]])
                print(f"    찾은 triplet: [{nums[i]}, {nums[left]}, {nums[right]}]")
                left += 1
                right -= 1
                # 중복 제거를 위한 추가 로직 생략 (설명 단순화)
                break
            elif total < 0:
                left += 1
            else:
                right -= 1
    
    print(f"결과: {result}")
    return result

def threeSum_skip_next(nums):
    """제안하신 방식: nums[i] == nums[i+1]인 경우 skip"""
    nums.sort()
    result = []
    n = len(nums)
    
    print(f"\n=== 방식 2: nums[i] == nums[i+1] ===")
    print(f"배열: {nums}")
    print("처리 순서:")
    
    for i in range(n - 2):
        # 경계 체크 필요!
        if i < n - 1 and nums[i] == nums[i + 1]:
            print(f"  i={i}: nums[{i}]={nums[i]} == nums[{i+1}]={nums[i+1]} → SKIP")
            continue
            
        print(f"  i={i}: nums[{i}]={nums[i]} → PROCESS")
        
        # 실제 triplet 찾기 (간단화)
        left, right = i + 1, n - 1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total == 0:
                result.append([nums[i], nums[left], nums[right]])
                print(f"    찾은 triplet: [{nums[i]}, {nums[left]}, {nums[right]}]")
                left += 1
                right -= 1
                break
            elif total < 0:
                left += 1
            else:
                right -= 1
    
    print(f"결과: {result}")
    return result

def demonstrate_difference():
    """두 방식의 차이점을 명확히 보여주는 예제"""
    
    # 예제 1: 연속된 중복값이 있는 경우
    print("=" * 60)
    print("예제 1: [-1, -1, 0, 1, 2] - 중복이 앞쪽에 있는 경우")
    print("=" * 60)
    
    nums1 = [-1, -1, 0, 1, 2]
    threeSum_skip_previous(nums1.copy())
    threeSum_skip_next(nums1.copy())
    
    # 예제 2: 중복이 여러 군데 있는 경우
    print("\n" + "=" * 60)
    print("예제 2: [-2, -1, -1, 0, 1, 1, 2] - 중복이 여러 곳에")
    print("=" * 60)
    
    nums2 = [-2, -1, -1, 0, 1, 1, 2]
    threeSum_skip_previous(nums2.copy())
    threeSum_skip_next(nums2.copy())

def explain_key_differences():
    """핵심 차이점 설명"""
    print("\n" + "=" * 60)
    print("핵심 차이점 분석")
    print("=" * 60)
    
    print("배열: [-1, -1, 0, 1, 2]")
    print("인덱스:  0   1  2  3  4")
    print()
    
    print("방식 1: if i > 0 and nums[i] == nums[i-1]:")
    print("  i=0: 첫 번째이므로 처리 (PROCESS)")
    print("  i=1: nums[1]=-1 == nums[0]=-1 → SKIP")
    print("  i=2: nums[2]=0 != nums[1]=-1 → PROCESS")
    print("  → 결과: 첫 번째 -1만 처리")
    print()
    
    print("방식 2: if nums[i] == nums[i+1]:")
    print("  i=0: nums[0]=-1 == nums[1]=-1 → SKIP")
    print("  i=1: nums[1]=-1 != nums[2]=0 → PROCESS")
    print("  i=2: nums[2]=0 != nums[3]=1 → PROCESS")
    print("  → 결과: 두 번째 -1만 처리")
    print()
    
    print("💡 차이점:")
    print("  • 방식 1: 중복 중 '첫 번째' 항목을 처리")
    print("  • 방식 2: 중복 중 '마지막' 항목을 처리")
    print()
    
    print("🚨 추가 문제점:")
    print("  • 방식 2는 배열 끝에서 out-of-bounds 위험")
    print("  • nums[i+1] 접근 시 i < n-1 체크 필요")

if __name__ == "__main__":
    demonstrate_difference()
    explain_key_differences()
```

## 168. Two Sum II

```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        left = 0; right = len(numbers) - 1
        while left < right:
            current_sum = numbers[left] + numbers[right]
            if current_sum == target:
                return [left + 1, right + 1]
            elif current_sum < target:
                left += 1
            else:
                right -= 1
        return []
```
